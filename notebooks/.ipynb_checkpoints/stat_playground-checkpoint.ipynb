```python
import pandas as pd #Creating dataframe and manipulating data
import re #regular expressions used for detection of initials
import seaborn as sns # data visualizations
from pathlib import Path
import numpy as np
from habanero import Crossref, WorksContainer
import py3langid as langid
```

```python
# Data Directory
data_dir = Path('../data')
input_dir = data_dir / 'input'
output_dir = data_dir / 'output'
# Loading in dataset
df = pd.read_csv(output_dir / '04_language_detection.csv').set_index('Index')
```

```python
df.lang_type.replace(0, 'English', inplace=True)
df.lang_type.replace(1, 'Monolingual (excl. en)', inplace=True)
df.lang_type.replace(2, 'Multilingual', inplace=True)
```

```python
df
```

```python
import pingouin as pg
completeness = ['no_author', 'no_language', 'no_title']
quality = ['author_sequence', 'author_initials', 'author_institutions',
              'author_characters', 'abstract_multi_lang', 'title_language']
df['completeness_errors'] = df[completeness].sum(axis=1)
df['quality_errors'] = df[quality].sum(axis=1)
pg.kruskal(data=df, dv='quality_errors', between='lang_type')
```

```python
expected, observed, stats = pg.chi2_independence(df, x='lang_type', y='quality_errors')
```

```python
expected.round(1)
```

```python
observed
```

```python
stats.round(2)
```

```python
pg.kruskal(data=df, dv='completeness_errors', between='lang_type')
```

```python
expected, observed, stats = pg.chi2_independence(df, x='lang_type', y='completeness_errors')
```

```python
expected.round(1)
```

```python
observed
```

```python
stats.round(2)
```

```python
grouped = df.groupby('lang_type')
aggie_land = grouped.agg({'DOI': pd.Series.nunique, 'total_errors' : 'sum', 'completeness_errors': 'sum' , 'quality_errors': 'sum'})
aggie_land.sort_values(by='DOI', ascending=False)
```
