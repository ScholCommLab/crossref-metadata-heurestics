```python
# Data Directory
data_dir = Path('../data')
input_dir = data_dir / 'input'
output_dir = data_dir / 'output'
# Loading in dataset
df = pd.read_csv(output_dir / '03_labeled_data.csv', 
                 usecols=['Index', 'publisher', 'container-title', 'DOI','indexed', 'created', 'title', 'abstract', 'author', 'language',
                         'original-title','no_author', 'no_language', 'no_abstract', 'no_title', 'author_sequence','author_initials','author_institutions',
                         'author_characters','abstract_multi_lang', 'title_language', 'total_errors'],
                 parse_dates=['created', 'indexed'], infer_datetime_format=True, 
                index_col=['Index'])
#df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df['created_year'] = df.created.dt.year
```

```python
pub_count = df.publisher.value_counts()
#pub_count.cut(bins=5)
cut_bins = pd.cut(pub_count, bins=[0,75,1000,5000,10000,20000], labels=['XS','S','M','L','XL'])
def bin_applicator(record):
    try:
        pub = record['publisher']
        locator = cut_bins.loc[cut_bins.index == pub]
        locList = locator.to_list()
        return locList[0]
    except:
        return None
df['publisher_bin'] = df.apply(bin_applicator, axis=1)
df['publisher_bin']
```
