```python
# Data Directory
data_dir = Path('../data')
input_dir = data_dir / 'input'
output_dir = data_dir / 'output'
# Loading in dataset
df = pd.read_csv(output_dir / '04_language_detection.csv', parse_dates=['created', 'deposited'],
                 infer_datetime_format=True, index_col = 'Index')
```

```python
completeness_errors = ['no_author', 'no_language', 'no_title']
quality_errors = ['author_sequence', 'author_initials', 'author_institutions', 'author_characters', 'abstract_multi_lang', 'title_language']
df['completeness_errors'] = df[completeness_errors].sum(axis=1)
df['quality_errors'] = df[quality_errors].sum(axis=1)
```

```python
import ast
def reformat_col(record):
    try:
        formed = ast.literal_eval(record)
        return formed
    except:
        return None

cols_to_reformat = ['detected_lang']
for col in cols_to_reformat:
    df[col] = df[col].apply(lambda x: reformat_col(x))
```
